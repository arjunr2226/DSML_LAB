```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("houseprice.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

features = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']

le = LabelEncoder()

for i in features:
    df[i] = le.fit_transform(df[i])
df
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=56)
X_train['area']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train.iloc[:, 0] = sc.fit_transform(X_train.iloc[:, 0].values.reshape(-1, 1))
X_test.iloc[:, 0] = sc.transform(X_test.iloc[:, 0].values.reshape(-1, 1))

X_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

mse = mean_squared_error(y_test, y_pred)

print(f"MSE {mse}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```
